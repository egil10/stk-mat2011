# Visualising the tick data

### Importing packages

```python
import pandas as pd
```

### Reading in data

```python
months = pd.date_range(start='2019-01-01', end='2026-03-01', freq='MS').strftime('%Y%m')

bids_list = []
asks_list = []

for m in months:
    bid = pd.read_parquet(f"../data/processed/eurusd_dukascopy_bid_{m}.parquet")
    ask = pd.read_parquet(f"../data/processed/eurusd_dukascopy_ask_{m}.parquet")

    bids_list.append(bid)
    asks_list.append(ask)

df_bid = pd.concat(bids_list, ignore_index=True)
df_ask = pd.concat(asks_list, ignore_index=True)

df_bid = df_bid.sort_values('datetime')
df_ask = df_ask.sort_values('datetime')

print(df_bid)
print(df_ask)
```
